```python
import pandas as pd

data = pd.read_csv('ficsgamesdb_2022_CvH_nomovetimes_304198 - Copy.txt', names = ['text'])
data
```

```python
data['heading'] = data['text'].str.split(' ').str[0].str.replace("[","")
data
```

```python
data['content'] = data['text'].apply(lambda st: st[st.find(' "')+2:st.find('"]')])
data
```

```python
import numpy as np
    
data['match'] = 0

for i in range (1,len(data)):
    if data.loc[i,'heading'] == 'Event':
        data.loc[i,'match'] = 1 + data.loc[i-1,'match']
    else:
        data.loc[i,'match'] = data.loc[i-1,'match']

data.head(50)
```

```python
cols = list(data.heading.unique())
cols
```

```python
data1 = data[data['match'].isin([0,1,2,3])].drop(['text'], axis = 1)
data1
```

```python
data1.pivot(index='match',columns='heading',values='content').reset_index()
```
